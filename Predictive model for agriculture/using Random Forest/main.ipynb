```python
# Libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import joblib
```

```python
# Load Dataset

crops = pd.read_csv("/content/ML projects/predictive model for agriculture/soil_measures.csv")

# Checking the first few rows

crops.head()
```

```python
# Features and Target

X = crops[["N", "P", "K", "ph"]]
y = crops["crop"]
```

```python
# Encode crop labels

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Split Dataset

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
# Train Model

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
# Train Model

model.fit(X_train, y_train)
```

```python
# Evaluate Model

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
# Save Model

joblib.dump(model, "crop_prediction_model.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")

print("\nModel saved successfully!")
```

```python
# Prediction

print("Crop Prediction")

N = float(input("Enter Nitrogen (N): "))
P = float(input("Enter Phosphorous (P): "))
K = float(input("Enter Potassium (K): "))
ph = float(input("Enter pH Value: "))

user_data = pd.DataFrame({
    "N": [N],
    "P": [P],
    "K": [K],
    "ph": [ph]
})

prediction = model.predict(user_data)

crop = label_encoder.inverse_transform(prediction)

probabilities = model.predict_proba(user_data)[0]
confidence = max(probabilities) * 100


print("Recommended Crop:", crop[0])
print(f"Confidence: {confidence:.2f}%")
```
